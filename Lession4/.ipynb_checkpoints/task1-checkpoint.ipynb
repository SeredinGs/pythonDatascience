С помощью функции plot постройте график, соединяющий линиями точки с горизонтальными координатами из списка x и вертикальными - из списка y.

Импортируем используемые библиотеки и зададим необходимые параметры

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Создаем списки х и у:

```python
x = [1,2,3,4,5,6,7]
y = [3.5, 3.8, 4.2, 4.5, 5.0, 5.5,7]
```

Построим график:

```python
plt.plot(x,y)
```

Построим scatter-график

```python
plt.scatter(x,y)
```

Также для построения 2 графиков в одном figure необходимо использовать функцию subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, y)
ax2.scatter(x,y)
```
